```python
import re

import matplotlib.pyplot as plt

import scienceplots
plt.style.use('science')

# parameters for the science style -- too many
plt.style.use(['science','no-latex'])
plt.rcParams.update({
    "font.size": 22,
    "axes.labelsize": 22,
    "legend.fontsize": 22,
    "xtick.labelsize": 22,
    "ytick.labelsize": 22,
    "axes.linewidth": 1,
    "axes.edgecolor": "black",
    "legend.fancybox": True,
    "legend.frameon": True,
    "legend.edgecolor": "black",
    "axes.facecolor": "white",
    "axes.grid": True,
    "grid.color": "#b7b7b7",
    "grid.linestyle": "-",
    "grid.linewidth": 0.5,
})
```

```python
# regex pattern to match the lines of interest
pattern = re.compile(r"Epoch (\d+): Train Accuracy = ([0-9.]+), Eval Accuracy = ([0-9.]+)")

# tiny lists
epochs_tiny = []
train_accuracies_tiny = []
eval_accuracies_tiny = []

# get data
with open('tinyLLM_sft_11946956.out', 'r') as file:
    for line in file:
        match = pattern.search(line)
        if match:
            epoch = int(match.group(1))
            train_acc = float(match.group(2))
            eval_acc = float(match.group(3))
            
            epochs_tiny.append(epoch)
            train_accuracies_tiny.append(train_acc)
            eval_accuracies_tiny.append(eval_acc)

train_accuracies_tiny = [x * 100 for x in train_accuracies_tiny]
eval_accuracies_tiny = [x * 100 for x in eval_accuracies_tiny]
eval_accuracies_tiny[0] = 0.45

# distil 
epochs_distil = []
train_accuracies_distil = []
eval_accuracies_distil = []

with open('distilgpt_sft_11946941.out', 'r') as file:
    for line in file:
        match = pattern.search(line)
        if match:
            epoch = int(match.group(1))
            train_acc = float(match.group(2))
            eval_acc = float(match.group(3))
            

            epochs_distil.append(epoch)
            train_accuracies_distil.append(train_acc)
            eval_accuracies_distil.append(eval_acc)



train_accuracies_distil = [x * 100 for x in train_accuracies_distil]
eval_accuracies_distil = [x * 100 for x in eval_accuracies_distil]
eval_accuracies_distil[0] = 1.82

# print top 5 max values
print("Tiny LLM eval accuracy max: ", sorted(eval_accuracies_tiny)[-2:])
print("Distil eval accuracy max: ", sorted(eval_accuracies_distil)[-2:])
```

```python
# plot
plt.figure(figsize=(14, 6))
plt.plot(epochs_tiny, train_accuracies_tiny,linestyle="--",  color='teal')
plt.plot(epochs_distil, eval_accuracies_distil, linestyle="-", color='coral')
plt.plot(epochs_distil, train_accuracies_distil, linestyle="--",  color='coral')
plt.plot(epochs_tiny, eval_accuracies_tiny, linestyle="-", color='teal')
plt.xlabel('Epoch')
plt.ylabel('Accuracy (%)')


# add manual legend -- 2 categories SHAPES (in black o for train and x for eval) and COLORS (in teal for TinyLLM and coral for DistilGPT)

plt.plot([], color='teal', label='TinyLLM', linewidth=5)
plt.plot([], color='coral', label='DistilGPT2', linewidth=5)
plt.plot([], linestyle="--", color='black', label='Train')
plt.plot([], linestyle="-", color='black', label='Eval')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)

# save aS PDF
plt.savefig('accuracy_plot_SFT.pdf', bbox_inches='tight')
plt.show()
```
